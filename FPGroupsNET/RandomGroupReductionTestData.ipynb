```python
import pandas as pd
import seaborn as sns
import json
import matplotlib.pyplot as plt

# Load JSON file
with open('/home/johannesh/repos/FPGroups/FPGroupsNET/RandomGroupReductionTestData.json') as f:
    data = json.load(f)

# Convert JSON data to DataFrame
df = pd.DataFrame(data)
# Filter dataframe to show only those with ProportionOfGenerators < 0.6
df2 = df[df['ProportionOfGenerators'] < 0.6]

# Pivot the DataFrame to get the format needed for the heatmap
pivot_df = df2.pivot_table(values='RelLengthOpt', index='VertexCount', columns='EdgeCount')

# Plot heatmap
plt.figure(figsize=(10,8))
sns.heatmap(pivot_df, cmap='viridis')

plt.show()
```
